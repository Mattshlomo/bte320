```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float))
  assert isinstance(annual_rate, (int, float))
  assert isinstance(months, int)

  if principal <= 0 or annual_rate < 0 or months <= 0:
      raise ValueError("All inputs must be positive")

  monthly_rate = annual_rate / 100 / 12

  if monthly_rate != 0:
      monthly_payment = (monthly_rate * principal) / (1 - (1 + monthly_rate) ** -months)
  else:
      monthly_payment = principal / months
  return monthly_payment
```

```python
def getValues(n):
  assert isinstance(n, int) # or: assert type(n) == int
  assert n > 0
  values = []

  for i in range(n):
    values.append(float(input("Enter a value: ")))

  if len(values) < 3:
    raise ValueError("The length of 'alist' must be greater than 2.")
  return values[2:]

def calcAverage(alist):
  assert isinstance(alist, list), "Input parameter 'alist' must be a list."
  assert len(alist) > 0, "Input parameter 'alist' must not be empty."

  try:
    return sum(alist) / len(alist)
  except Exception as e:
    raise ValueError(f"error: {e}")



def main():
  n = int(input("Enter number of items to be added to the list: "))
  alist = getValues(n)
  avg = calcAverage(alist)
  print(avg)

main()
```

```python
class Greeting:

  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal = False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return "Instance of class 'Greeting'"

name =  Greeting("John", "Doe")
print(name)
name.display()
name.display(formal=True)
```

```python
class Student:
  def __init__(self,name,number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newname):
    assert isinstance(newname, str)
    self.__name = newname

  def __len__(self):
    return len(self.__courses)

  def enroll(self, addcourse):
    if addcourse not in self.__courses:
      self.__courses.append(addcourse)
    else:
      print(f"Student {self.__name} is already enrolled in {addcourse}.")
```

```python
s = Student("Jane Doe", 12345)
print(f"Student name:{s.get_name()}")
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Classes already enrolled in: {s.get_courses()}")
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5

class Rectangle(Point):
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(5,0)
    self.p3 = Point(0,5)
    self.p4 = Point(5,5)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height
```

```python
z = Rectangle()
z.area()
```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__wage = wage
    self.hours = hours

  def calculate_pay(self):
    return self.__wage * self.hours

  def __str__(self):
    return f"Name: {self.name}\n Wage: {self.__wage}\n Hours: {self.hours}\n"

  def get_wage(self):
    return self.__wage

  def set_wage(self, newwage):
    if newwage <= 0:
      print("Invalid wage value")
    else:
      self.__wage = newwage


class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus #deciaml format

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours


employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly payment is {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly payment is {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus.")
```
